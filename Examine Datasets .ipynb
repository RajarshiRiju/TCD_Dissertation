```python
#Group Plaza IDs based on Toll
def addTollName(plazaid):
    if plazaid in (1416,1417):
        return "Limerick Tunnel"
    elif plazaid in (1415,2885,2886):    # 0 added as dummy ---- TO-DO
        return "M1"
    elif plazaid in (2862,2861):
        return "M3"
    elif plazaid in (2863,2864,2865,2866):
        return "M4"
    elif plazaid in (1420,1421):
        return "M50"
    elif plazaid in (1419,0):
        return "M7-M8"
    elif plazaid in (1423,0):
        return "Dublin Tunnel"
    elif plazaid in (1424,0):
        return "Eastlink"
    elif plazaid in (1422,0):
        return "N25"
    elif plazaid in (1418,0):
        return "N6"
    elif plazaid in (2869,2870):
        return "M8"
    else:
        return "--"
    
```

## Read Counter Dataset

```python
ctr = pd.read_csv('F:/counterData/2019-04-09.csv')
```

```python
ctr_s = ctr[(ctr["cosit"]==1508) & (ctr["lanename"].str.startswith("Southbound 1"))]
```

```python
ctr_s.head()
```

```python
ctr_s.sort_values(by=["hour", "minute", "second", "millisecond"], inplace=True)
```

```python
tab = pd.DataFrame()
tab["timegap"] = (ctr_s["hour"]*3600 + ctr_s["minute"]*60 + ctr_s["second"]).diff()
tab["gap"] = ctr_s["gap"]
tab["speed"] = ctr_s["speed"]
```

```python
tab.sort_values(by=["timegap"], inplace=True)
```

```python
fig = plt.figure()
#ax = fig.add_axes([0,0,1,1])
plt.figure(figsize=(10,5)) 
plt.scatter(tab['timegap'][:-3], tab['gap'][:-3])
plt.show()
```

```python
#fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
#ax1.figure(figsize=(10,5)) 
ax1.bar(tab['timegap'][:-3], tab['speed'][:-3])
#ax2.figure(figsize=(10,5)) 
ax2.scatter(tab['timegap'][:-3], tab['speed'][:-3].diff())
plt.show()
```

```python
#E:\Trinity\Dissertation\Sem 3\Weather
weather = pd.read_csv('E:/Trinity/Dissertation/Sem 3/Weather/Dublin Jan 2018.csv')
```

```python
weather = weather[8760:8784]  #Taking only January for now
weather["date"] = weather["date"].str[11:13]
weather.rename(columns={'date': 'hour'}, inplace=True)
```

```python
weather = weather.apply(pd.to_numeric)
```

```python
ctr = pd.read_csv('F:/counterData/2019-01-01.csv')
```

```python
ctr = ctr[ctr["cosit"]==1508]
Freq_df = ctr.groupby(['hour']).size().reset_index(name='Freq')
```

```python
plt.figure(figsize=(10,7))
corr = weather.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    linewidths=0.01, 
    linecolor='black',
    annot=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```
